## 3.2. Deletions and additions Example

```kotlin
val wA = HC.importXMLWitness("A", """
            <s>&amp; <add>finally</add> in what order he would eat the biscuits<del>, in the cockpit</del> when the time came.</s>
            """.trimMargin())
val wB = HC.importXMLWitness("B", """
            |<s>&amp; <app>
            |<rdg varSeq="1"/>
            |<rdg varSeq="2">finally</rdg>
            |</app> in what order he would eat the biscuits<app>
            |<rdg varSeq="1">, in the cockpit</rdg>
            |<rdg varSeq="2"/>
            |</app> when the time came.</s>
            """.trimMargin())
wA.show()
```

```kotlin
wB.show()
```

```kotlin
val wC = HC.importXMLWitness("C","<s>&amp; finally in what order he would eat the biscuits when the time came.</s>")
val cg = HC.collate(wA,wC)
cg.show()
```

```kotlin
cg.asASCIITable()
```

```kotlin
cg.asHtml()
```
